```python
import json
import math
import numpy as np

id_img_dict = {}
with open('annotations_train.json', 'r') as f:
    content = json.load(f)
    images = content["images"]
    anns = content["annotations"]
    for im in images:
        id_img_dict[im['id']] = im['file_name']
    
def dist(x1, y1, x2, y2):
    return math.sqrt((x1-x2)**2 + (y1-y2)**2) 

for ann in anns:
    segs = ann['segmentation']
    img_id = ann['image_id']
    for seg in segs:
        prev_x = 0
        prev_y = 0
        angles = []
        close_angles = []
        for i in range(0, len(seg), 2):
            x = seg[i]
            y = seg[i+1]
            
            if(i == 0 or dist(x, y, prev_x, prev_y) <= 25):
                prev_x = x
                prev_y = y
                continue
            
            dx = x - prev_x
            dy = y - prev_y
            
            angle = math.atan2(dy, dx) % (math.pi / 2)
            angles.append(angle)
            if(i == 2):
                close_angles.append(angle)
            else:
                avg = np.average(np.array(close_angles))
                # This is for the cases like when 89 degree and 0 degree are together.
                if(abs(angle - avg) > 0.2):
                    if(abs(angle + math.pi / 2 - avg) < 0.2):
                        close_angles.append(angle + math.pi / 2)
                    elif(abs(angle - math.pi / 2 - avg) < 0.2):
                        close_angles.append(angle - math.pi / 2)
                else:
                    close_angles.append(angle)
            
            prev_x = x
            prev_y = y
        
        avg_angles = np.average(np.array(close_angles))
        print(id_img_dict[img_id],(avg_angles / math.pi * 180) % 90)
            
        
    
```
